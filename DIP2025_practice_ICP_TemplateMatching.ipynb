```python
pad_task1 = cv2.imread('input/binary_task1.bmp', cv2.IMREAD_GRAYSCALE)

# Find the pad and probemark contours and draw.

"""
Your code here
eg:

"""
_, binary = cv2.threshold(
    pad_task1, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

contours, _ = cv2.findContours(
    binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

result = cv2.cvtColor(pad_task1, cv2.COLOR_GRAY2BGR)
pad_area = probemark_area = 0

for i in contours:
    area = cv2.contourArea(i)

    if area > 5000:                 # pad 視實際圖片大小調參
        pad_area = area
        cv2.drawContours(result, [i], -1, (0, 0, 255), 1)

    elif 20 < area < 300:           # probe mark  視實際圖片大小調參
        probemark_area += area
        cv2.drawContours(result, [i], -1, (0, 255, 0), 1)
cv2.imwrite("output/pad_task1_cont.bmp", result)
print('The pad area is ', pad_area)
print('The probe mark area is ', probemark_area)
```

## 7. ICP: Apple Edge Alignment

ICP moves one edge until it overlaps another edge.  
Here, the **blue apple** is the source and the **orange apple** is the target.

The apple edge is simply a list of $(x,y)$ points. The apple equation in the code only creates the example shape; it is not part of ICP.

### Step 1: Find the nearest point

For every blue point, ICP finds the nearest orange point. The distance between two points is:

$$
d=\sqrt{(x_1-x_2)^2+(y_1-y_2)^2}
$$

A smaller $d$ means the two points are closer.

### Step 2: Rotate and move the source

ICP calculates an angle $\theta$ and movement $(t_x,t_y)$ that make the matched points closer:

$$
x_{new}=x\cos\theta-y\sin\theta+t_x
$$

$$
y_{new}=x\sin\theta+y\cos\theta+t_y
$$

The code uses SVD to find the angle automatically; the SVD derivation is not required here.

### Step 3: Check the average error

$$
E=\frac{d_1+d_2+\cdots+d_N}{N}
$$

### Step 4: Repeat

ICP repeats **find nearest points → rotate and move → check error** until $E$ almost stops changing.

```python
from pathlib import Path

# 1. Create an apple-shaped edge
theta = np.linspace(0, 2 * np.pi, 180, endpoint=False)
radius = 1 + 0.32 * np.cos(2 * theta) - 0.08 * np.sin(theta)
apple_body = np.column_stack((radius * np.cos(theta),
                              1.05 * radius * np.sin(theta)))

# Add a stem to make the shape less symmetric
apple_stem = np.column_stack((np.linspace(0.00, 0.15, 20),
                              np.linspace(0.62, 1.20, 20)))

# Add an elliptical leaf
u = np.linspace(0, 2 * np.pi, 50, endpoint=False)
leaf_angle = 0.45
leaf_x = (0.18 + 0.28 * np.cos(u) * np.cos(leaf_angle)
          - 0.09 * np.sin(u) * np.sin(leaf_angle))
leaf_y = (1.06 + 0.28 * np.cos(u) * np.sin(leaf_angle)
          + 0.09 * np.sin(u) * np.cos(leaf_angle))
apple_leaf = np.column_stack((leaf_x, leaf_y))

source = np.vstack((apple_body, apple_stem, apple_leaf))

# 2. Create a rotated and shifted target
angle = np.deg2rad(12)
R_true = np.array([[np.cos(angle), -np.sin(angle)],
                   [np.sin(angle),  np.cos(angle)]])
target = source @ R_true.T + np.array([0.35, 0.15])

# 3. ICP alignment
aligned = source.copy()
previous_error = np.inf

for iteration in range(7):
    # Find nearest target point
    distances = np.linalg.norm(aligned[:, None, :] - target[None, :, :], axis=2)
    nearest = target[np.argmin(distances, axis=1)]

    # Remove the centers
    center_p = aligned.mean(axis=0)
    center_q = nearest.mean(axis=0)
    P = aligned - center_p
    Q = nearest - center_q

    # SVD gives the best rotation
    H = P.T @ Q
    U, _, Vt = np.linalg.svd(H)
    R = Vt.T @ U.T
    if np.linalg.det(R) < 0:
        Vt[-1, :] *= -1
        R = Vt.T @ U.T

    # Translation and point update
    t = center_q - center_p @ R.T
    aligned = aligned @ R.T + t

    error = np.mean(np.linalg.norm(aligned - nearest, axis=1))
    if abs(previous_error - error) < 1e-6:
        break
    previous_error = error

# 4. Show before and after ICP
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].scatter(source[:, 0], source[:, 1], s=8, label='Source')
ax[0].scatter(target[:, 0], target[:, 1], s=8, label='Target')
ax[0].set_title('Before ICP')
ax[0].axis('equal')
ax[0].legend()

ax[1].scatter(aligned[:, 0], aligned[:, 1], s=8, label='Aligned')
ax[1].scatter(target[:, 0], target[:, 1], s=8,
              facecolors='none', edgecolors='red', label='Target')
ax[1].set_title('After ICP')
ax[1].axis('equal')
ax[1].legend()

Path('output').mkdir(exist_ok=True)
plt.tight_layout()
plt.savefig('output/icp_result.png')
plt.show()

print('Iterations:', iteration + 1)
print('Mean alignment error:', error)
```
